```python
df.isnull() .sum()
```

```python
df["Gender"].fillna("No Gender", inplace = True)
df.isnull().sum()
```

```python
df["Seniority level"].fillna("Seniority level  ", inplace = True)
df.isnull().sum()
```

```python
df["Your main technology / programming language"].fillna("Your main technology / programming language", inplace = True)
df.isnull().sum()
```

```python
df["Other technologies/programming languages you use often"].fillna("Other technologies/programming languages you use often", inplace = True)
df.isnull().sum()
```

```python
df["Employment status"].fillna("Employment status", inplace = True)
df.isnull().sum()
```

```python
df["Сontract duration"].fillna("Сontract duration", inplace = True)
df.isnull().sum()
```

```python
df["Main language at work"].fillna("Main language at work", inplace = True)
df.isnull().sum()
```

```python
df["Have you lost your job due to the coronavirus outbreak?"].fillna("Have you lost your job due to the coronavirus outbreak?", inplace = True)
df.isnull().sum()
```

```python
df["Company type"].fillna("Company type", inplace = True)
df.isnull().sum()
```

```python
df["Age"].fillna(int(df["Age"].mean()), inplace=True)
df.isnull().sum()
```

```python

df = df.dropna(axis = 0, how ='any')
  
print(df.isnull().sum())
df.shape
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
  
  
sns.histplot(x='Сontract duration', data=df, )
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
  
```

```python
df['Number of vacation days'] = pd.to_numeric(df['Number of vacation days'],errors='coerce').fillna(0).astype(int)
df['Years of experience in Germany'] = pd.to_numeric(df['Years of experience in Germany'],errors='coerce').fillna(0).astype(int)
df['Yearly brutto salary (without bonus and stocks) in EUR'] = pd.to_numeric(df['Yearly brutto salary (without bonus and stocks) in EUR'],errors='coerce').fillna(0).astype(int)
df['Age'] = df['Age'].fillna(0).astype(int)
```

```python
fig =plt.figure(figsize=(14,7))
plt.title('Value distribution:', fontsize= 17)
axs = fig.subplots(nrows=2, ncols=2)
sns.boxplot(data = df, x='Age', ax = axs[0,0])
sns.boxplot(data = df, x='Years of experience in Germany', ax = axs[0,1])
sns.boxplot(data = df, x='Number of vacation days', ax = axs[1,1])
sns.boxplot(data = df, x='Yearly brutto salary (without bonus and stocks) in EUR', ax = axs[1,0])
plt.show()
```
